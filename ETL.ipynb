```python
import pandas as pd
```

```python
df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
```

```python
display(df.shape)
display(df.head(10))
```

```python
# type of ship modes
df['Ship Mode'].unique()
```

```python
# cleaning the column names

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
#df.head(10)
```

```python
# Defining new column to analyse the total amount of dicount given
df['discount'] = df['list_price']*df['discount_percent']/100

# Definig new column to evaluate the final price after applying discounts
df['sale_price'] = df['list_price'] - df['discount']

# Defining new column to evaluate the total profits
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
df.head(10)
```

```python
# Data-type check
display(df.dtypes)
```

```python
# Converting date data-type
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")

display(df.dtypes)
```

```python
# Removing unecessary columns

df.drop(columns=['list_price', 'discount_percent', 'cost_price'], inplace = True)
```

```python
df.head()
```

```python
df.to_csv("data.csv", index=False)
```
